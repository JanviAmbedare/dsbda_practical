```python
def impute_age(cols): #designed to fill in missing age values in a dataset
    Age = cols[0]
    Pclass = cols[1]
    
    if pd.isnull(Age):
        if Pclass ==1:
            return 37
        elif Pclass ==2:
            return 29
        else:
            return 24
    else:
        return Age
            
```

```python
df['Age'] = df[['Age', 'Pclass']].apply(impute_age, axis=1)
#This will fill in the missing values in the 'Age' column using the logic in impute_age
df.drop('Cabin', axis=1, inplace=True)
df.dropna(inplace=True)
```
